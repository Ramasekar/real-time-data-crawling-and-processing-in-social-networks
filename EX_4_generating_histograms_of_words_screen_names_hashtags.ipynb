```
import twitter
```

```
CONSUMER_KEY = "+++++++++++++++++++++++++++++++++"
CONSUMER_SECRET = "++++++++++++++++++++++++++++++++++++++++"

OAUTH_TOKEN = "+++++++++++++++++++++++++++++++++"
OAUTH_TOKEN_SECRET = "++++++++++++++++++++++++++++++++++++++++++""
```

```
auth=twitter.oauth.OAuth(OAUTH_TOKEN,OAUTH_TOKEN_SECRET,CONSUMER_KEY,CONSUMER_SECRET)
twitter_api=twitter.Twitter(auth=auth)
```

```
trend_tweet='#TuesdayThoughts'
count=50
search_result=twitter_api.search.tweets(q=trend_tweet,count=count)
print(search_result)
```

```
statuses=search_result['statuses']
```

```
status_texts = [ status['text'] 
                 for status in statuses ]
```

```
status_texts
```

```
screen_names = [ user_mention['screen_name'] 
                 for status in statuses
                     for user_mention in status['entities']['user_mentions'] ]
```

```
screen_names
```

```
hashtags = [ hashtag['text'] 
             for status in statuses
                 for hashtag in status['entities']['hashtags'] ]
```

```
hashtags
```

```
text = [ w 
          for t in status_texts 
              for w in t.split() ]

text
```

```
from collections import Counter
import matplotlib.pyplot as plt
```

```
counter = Counter(text)
```

```
counter
```

```
#text
plt.title("Text")
plt.hist(list(counter.values()))
plt.xlabel("NO. of times text appeared")
plt.ylabel("No. of items in the text")
```

```
counter_hashtags = Counter(hashtags)
counter_hashtags
```

```
#hashtags
plt.title("HASHTAGS")
plt.hist(list(counter_hashtags.values()))
plt.xlabel("NO. of times text appeared")
plt.ylabel("No. of items in the text")
```

```
#screen names
counter_screen_names = Counter(screen_names)
counter_screen_names
```

```
#screen names
plt.title("SCREEN NAMES")
plt.hist(list(counter_screen_names.values()))
plt.xlabel("NO. of times text appeared")
plt.ylabel("No. of items in the text")
```

```
#text
plt.title("Text")
plt.hist(list(counter.values()))
plt.xlabel("NO. of times text appeared")
plt.ylabel("No. of items in the text")


#hashtags
plt.title("HASHTAGS")
plt.hist(list(counter_hashtags.values()))
plt.xlabel("NO. of times text appeared")
plt.ylabel("No. of items in the text")


#screen names
plt.title("SCREEN NAMES, TEXT, HASHTAGS")
plt.hist(list(counter_screen_names.values()))
plt.xlabel("NO. of times text appeared")
plt.ylabel("No. of items in the text")
```
